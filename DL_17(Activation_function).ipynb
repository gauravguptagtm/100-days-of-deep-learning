# Activation Functions?

- The activation function decides whether a neuron should be activated or not by calculating the weighted sum and further adding bias to it. The purpose of the activation function is to introduce non-linearity into the output of a neuron.

$$
a = g(w1+w2+..+b)
$$

- A neural network without an activation function is essentially just a linear regression model. The activation function does the non-linear transformation to the input making it capable to learn and perform more complex tasks.


# Ideal Activation Function

- Activation function should be non-linear.
- It should be differentiable because of GD(not mandatory).
- It should bw computationally inexpensive.
- Zero Centered(Normalized)
- Non-saturating (saturating(in range) -> vanishing gradient proble)

## Sigmoid Activation function

image.png

$$
A = \frac{1}{1 + e^{-x}}
$$

**Advantages**
- Used in binary classification in output
- Non-linear function
- It's differentiable

**Disadvantages**
- Saturating Function -> Vanishing Gradiant Problem
- Non zero centered -> training slow
- Computationally expensive


## Tanh Activation function

image.png

** Advantages **
- Non-linear
- Differentiable
- Zero centered

**Disadvantes**
- Saturating function
- computationally expensive


## ReLu Activation function

image.png

$$
A(x) = max(0,x)
$$

**Advantages**

- Non-linear
- Not saturated in +ve region
- computationally inexpensive
- convergence faster than sigmoid and tanh

**Disadvantages**

- Not Differentiable at 0
- Non-zero centered(Solved by batch normalization)

## Dying Relu Problem
- Ouptut become zero in some neuron(Dead Neuron)
- if more than 50% neuron become dead neuron, then we can't capture representation.
- Main reasons:
  - high learning rate
  - high -ve bias

- Solutions
  - set low learning rate
  - bias -> +ve value (0.01)
  - use variant of relu

# Variants of RELU

- Linear variants

- Non-linear Variants
  - ELU (exponential linear unit)
