# Backpropogation

- Simply say that a algo to train a NN.(finding the best weight and value)
- Prerequisite:
    - Gradiant Descent
    - Forward Propogation

Steps:
- Intialize weights and biases.
- We select a point(row).
- Predict -> forward propogation[Dot product]
- We choose a loss function and calculate the loss.
- Update weight and bias using Gradiant Descent.

$$
W_{new} = W_{old} - lr*\frac{δL}{δW_{old}}
$$
- calculate avg loss for that epoch.


Note: Gradiant is parial deravative of Loss function.



# Regression Problem

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame([[8,8,4],[7,9,5],[6,10,6],[5,12,7]], columns=['cgpa', 'profile_score', 'lpa'])
```

```python
df
```

```python
def initialize_parameter(layer_dim):
  np.random.seed(3)
  parameters = {}
  L = len(layer_dim)

  for i in range(1, L):
    parameters['W'+str(i)] = np.ones((layer_dim[i-1], layer_dim[i]))*0.1
    parameters['b'+str(i)] = np.zeros(( layer_dim[i], 1))

  return parameters
```

```python
initialize_parameter([2,2,1])
```

```python
def linear_forward(A_prev, W, b):
  Z = np.dot(W.T, A_prev) + b
  return Z
```

```python
def L_layer_forward(X, parameters):
  A = X
  L = len(parameters) //2

  for i in range(1, L+1):
    A_prev = A
    w1 = parameters['W'+str(i)]
    b1 = parameters['b'+str(i)]
    print("A"+str(i-1)+": ", A_prev)
    print("W"+str(i)+": ", w1)
    print("b"+str(i)+": ", b1)
    print("--" *20)

    A = linear_forward(A_prev, w1, b1)
    print("A"+str(i)+": ", A)
    print("*"*20)

  return A, A_prev
```

```python
X = df[['cgpa', 'profile_score']].values[0].reshape(2,1) # Shape(no of features, no. of training example)
y = df[['lpa']].values[0][0]

# Parameter initialization
parameters = initialize_parameter([2,2,1])

y_hat,A1 = L_layer_forward(X, parameters)
```

```python
y_hat = y_hat[0][0]
```

```python
def update_parameters(parameters,y,y_hat,A1,X):
  parameters['W2'][0][0] = parameters['W2'][0][0] + (0.001 * 2 * (y - y_hat)*A1[0][0])
  parameters['W2'][1][0] = parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat)*A1[1][0])
  parameters['b2'][0][0] = parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat))

  parameters['W1'][0][0] = parameters['W1'][0][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0]*X[0][0])
  parameters['W1'][0][1] = parameters['W1'][0][1] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0]*X[1][0])
  parameters['b1'][0][0] = parameters['b1'][0][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0])

  parameters['W1'][1][0] = parameters['W1'][1][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0]*X[0][0])
  parameters['W1'][1][1] = parameters['W1'][1][1] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0]*X[1][0])
  parameters['b1'][1][0] = parameters['b1'][1][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0])
```

```python
update_parameters(parameters,y,y_hat,A1,X)
```

```python
parameters
```

```python
# epochs implementation

parameters = initialize_parameter([2,2,1])
epochs = 5

for i in range(epochs):

  Loss = []

  for j in range(df.shape[0]):

    X = df[['cgpa', 'profile_score']].values[j].reshape(2,1) # Shape(no of features, no. of training example)
    y = df[['lpa']].values[j][0]

    # Parameter initialization


    y_hat,A1 = L_layer_forward(X,parameters)
    y_hat = y_hat[0][0]

    update_parameters(parameters,y,y_hat,A1,X)

    Loss.append((y-y_hat)**2)

  print('Epoch - ',i+1,'Loss - ',np.array(Loss).mean())

parameters
```

#Intution behind the Backpropogtion

- Loss function is a function of all trainable parameters.
- Concept of Gradient
  - it's just derivative.
  - Difference is if depend on single parameter called derivative but if function depend on multiple parameter called gradient.
- Concept of Derivative
  - rate of change
- concept of Minima
  - where slope = 0, minima is there.

Assume lr = 1
$$
W_{new} = W_{old} - \frac{δL}{δW_{old}}
$$
